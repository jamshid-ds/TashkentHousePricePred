```python
import pandas as pd #import pandas
import numpy as np
df = pd.read_csv('https://raw.githubusercontent.com/jamshid-ds/praktikum_datasets/main/housing_LR.csv', index_col=0)
df.head(10)
```

```python
import matplotlib.pyplot as plt #matplotlib and seaborn for visualization
import seaborn as sns
%matplotlib inline
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='size', y='price') #i wanted to see corrilation between 'size' and 'price
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data=df, x='size', y='price', line_kws={"color": "red"}) #let's see with regplot
plt.show()
```

```python
from sklearn.model_selection import train_test_split
train_set, test_set = train_test_split(df, test_size=0.10, random_state=2) # divided into train and test sets
```

# I see a big correlation between 'size' and 'price' here

```python
X_train = np.asanyarray(train_set[['size']])
Y_train = np.asanyarray(train_set[['price']])
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(X_train, Y_train)
```

```python
X_test = np.asanyarray(test_set[['size']])
Y_test = np.asanyarray(test_set[['price']])
```

```python
Y_predict = model.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error

MAE = mean_absolute_error(Y_test, Y_predict)
RMSE = np.sqrt(mean_squared_error(Y_test, Y_predict))

print(f"{MAE}")
print(f"{RMSE}")
```

```python
import joblib
joblib.dump(model, 'LR_model.joblib')
```
